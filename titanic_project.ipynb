```python
# female passengers who were in first class
titanic[(titanic["sex"] == "female") & (titanic["pclass"] == 1)]
```

```python
# female passengers who survived
titanic[(titanic["sex"] == "female") & (titanic["survived"] == 1)]
```

```python
# female passengers who survived and were in first class
titanic[(titanic["sex"] == "female") & (titanic["survived"] == 1) & (titanic["pclass"] == 1)]
```

```python
# female passengers that were in first class and did not survived
titanic[(titanic["sex"] == "female") & (titanic["survived"] != 1) & (titanic["pclass"] == 1)]
```

```python
titanic[titanic["age"] < "10" ]
```

```python
# child younger than 10 yo that survived
titanic[(titanic["age"] < "10") & (titanic["survived"] == 1) ]
```

```python
# add new column to show how many relatives did every passengers have in the titanic
titanic["total_relatives"] = titanic["sibsp"] + titanic["parch"]
titanic
```

```python
# top 20 survived passengers that had the most total relatives 
titanic[titanic["survived"] == 1].sort_values("total_relatives" , ascending=False).head(20)
```

```python
# top 20 not survived passengers that had the most total relatives 
titanic[titanic["survived"] == 0].sort_values("total_relatives" , ascending=False).head(20)
```

```python
# all the passengers that traveled alone by titanic
titanic[titanic["total_relatives"] == 0]
```

```python
# how many of alone passengers were men or women?
titanic[titanic["total_relatives"] == 0].value_counts("sex")
```

```python
# how many of alone passengers survived or died?
titanic[titanic["total_relatives"] == 0].value_counts("survived")
```

```python
# survived male that traveled alone
titanic[(titanic["total_relatives"] == 0)  & (titanic["sex"] == "male") & (titanic["survived"] == 1)]
```

```python
# survived female that traveled alone
titanic[(titanic["total_relatives"] == 0)  & (titanic["sex"] == "female") & (titanic["survived"] == 1)]
```
